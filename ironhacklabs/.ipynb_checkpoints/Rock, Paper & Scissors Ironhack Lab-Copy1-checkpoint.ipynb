```python
# Rock, Paper & Scissors - 6 

print("Welcome to the Rock, Paper and Scissors game!!!\n")

name = input("What is your name player 1?\n")

print("Hello", name, "!", "You are going to play against the Computer\n")

print("The battle has 7 rounds , the winner needs to win 4 times OR to have more victories than the opponent\n")

print("Lets start it ! Select rock, paper or scissors, DONT FORGET TO WRITE CORRECTLY\n")
           
gestures = ['rock' , 'paper' , 'scissors']
   
n_rounds = 0
rounds_to_win = 4
cpu_score = 0
player_score = 0   
    
while n_rounds < 7:
    
    options = input("Player 1 , what is your option? rock, paper or scissors?\n")

    if options == 'rock' or options == 'paper' or options == 'scissors':
        print("Ok, your option is:", options)
        
    elif options != 'rock' or options != 'paper' or options != 'scissors':
        while options != 'rock' or options != 'paper' or options != 'scissors': 
            print("Ops, we dont have this option...try again writing correctly\n") 
            options = input("Player_1, what is your option?\n")
            if options == 'rock' or options == 'paper' or options == 'scissors':
                print("Ok, your option is:", options)
                break      
            
    import random
    cpu = random.choice(gestures)
    print("The computer selected:", cpu, "\n")
    print("Player 1 -", options, "x", cpu, "- Computer")
     
    if options == cpu:
        n_rounds += 1
        print("Round", n_rounds, ":", "TIE!!! 0 POINTS FOR THE PLAYERS\n")
        print("------------------------------------------------------------")
         
    elif options != cpu:
        n_rounds += 1 

        if options == 'rock' and cpu == 'paper':
        
            print("Paper covers rock! ", "Round", n_rounds, ":", "The computer won this round!!!")
            cpu_score += 1
            print("The computer score is:", cpu_score, "\n")
            print("------------------------------------------------------------")
            if cpu_score == rounds_to_win:
                print("Stop! The computer achieved 4 wins, its the CHAMPION!")
                break

        elif options == 'rock' and cpu == 'scissors':
        
            print("Rock smashes scissors! ", "Round", n_rounds, ":", name, "won this round!!!")
            player_score += 1
            print(name, "score is:", player_score, "\n")
            print("------------------------------------------------------------")
            if player_score == rounds_to_win:
                print("Stop!", name, "achieved 4 wins, here we have the CHAMPION!")
                break
        
        elif options == 'paper' and cpu == 'rock':
        
            print("Paper covers rock! ", "Round", n_rounds, ":", name, "won this round!!!")
            player_score += 1
            print(name, "score is:", player_score, "\n")
            print("------------------------------------------------------------")
            if player_score == rounds_to_win:
                print("Stop!", name, "achieved 4 wins, here we have the CHAMPION!")
                break
        
        elif options == 'paper' and cpu == 'scissors':
        
            print("Scissors cuts paper! ", "Round", n_rounds, ":", "The computer won this round!!!")
            cpu_score += 1 
            print("The computer score is:", cpu_score, "\n")
            print("------------------------------------------------------------")
            if cpu_score == rounds_to_win:
                print("Stop! The computer achieved 4 wins, its the CHAMPION!")
                break
        
        elif options == 'scissors' and cpu == 'paper':
        
            print("Scissors cuts paper! ", "Round", n_rounds, ":", name, "won this round!!!")
            player_score += 1
            print(name, "score is:", player_score, "\n")
            print("------------------------------------------------------------")
            if player_score == rounds_to_win:
                print("Stop!", name, "achieved 4 wins, here we have the CHAMPION!")
                break
        
        elif options == 'scissors' and cpu == 'rock':
        
            print("Rock smashes scissors! ", "Round", n_rounds, ":", "The computer won this round!!!")
            cpu_score += 1
            print("The computer score is:", cpu_score, "\n")
            print("------------------------------------------------------------")
            if cpu_score == rounds_to_win:
                print("Stop! The computer achieved 4 wins, its the CHAMPION!")
                break

    if n_rounds == 7:
        if cpu_score < rounds_to_win and player_score < rounds_to_win:
            if player_score > cpu_score:
                print(name, "is the CHAMPION with more scores!!!")
            elif cpu_score > player_score: 
                print("The computer is the CHAMPION with more scores!!!")
            else:
                print("We dont have a champion today...")
        
```
