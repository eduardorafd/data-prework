```python
# Sorcerers wins if he succeeds in winning 3 spell clashes in a row.
POWER = {
    'Fireball': 50, 
    'Lightning bolt': 40, 
    'Magic arrow': 10, 
    'Black Tentacles': 25, 
    'Contagion': 45
}

gandalf = ['Fireball', 'Lightning bolt', 'Lightning bolt', 'Magic arrow', 'Fireball', 
           'Magic arrow', 'Lightning bolt', 'Fireball', 'Fireball', 'Fireball']

saruman = ['Contagion', 'Contagion', 'Black Tentacles', 'Fireball', 'Black Tentacles', 
           'Lightning bolt', 'Magic arrow', 'Contagion', 'Magic arrow', 'Magic arrow']

# Lists of spell powers -------------------------------------------------------------------------------

gandalf_power = []
saruman_power = []

for i in gandalf:
    gandalf_power.append(POWER[i])
for x in saruman:
    saruman_power.append(POWER[x])
    
print("Here we have their spell powers:\n")    
print("gandalf_power =", gandalf_power)
print("saruman_power =", saruman_power, "\n")

# I made the lists with the command above -------------------------------------------------------------------------------------

gandalf_power = [50, 40, 40, 10, 50, 10, 40, 50, 50, 50]
saruman_power = [45, 45, 25, 50, 25, 40, 10, 45, 10, 10]

spells = len(gandalf_power) + len(saruman_power)
print("The number of spells in the whole battle is:", spells, "\n")

# Clash ------------------------------------------------------------------------------------------------------------------------

gandalf_wins = 0
saruman_wins = 0
clash = 0

print("The battle starts now!!!\n")

for i in range(len(gandalf_power)):
    print("Gandalf", gandalf_power[i], "x", saruman_power[i], "Saruman")
    clash += 1
    if gandalf_power[i] > saruman_power[i]:
        print("Gandalf won the clash", clash, "\n")
        gandalf_wins += 1
        if gandalf_wins - saruman_wins == 3:
            print("STOP! We dont need the last clash because Gandalf won 3 times in a row")
            print("Gandalf won the battle!")
            print("But...if we had the last match , Gandalf would win by 50 x 10\n")
            break
    else:
        print("Saruman won the clash", clash, "\n")
        saruman_wins += 1
                                   
print("Gandalf won:", gandalf_wins, "clashs") 
print("Saruman won:", saruman_wins, "clashs")

if gandalf_wins > saruman_wins:
    print("Gandalf is the Champion again!")
else:
    print("Saruman is the Champion!")   
    
```

```python
# Average spell power of Gandalf and Saruman.

import statistics 

gandalf_power = [50, 40, 40, 10, 50, 10, 40, 50, 50, 50]
saruman_power = [45, 45, 25, 50, 25, 40, 10, 45, 10, 10]

x = statistics.mean(gandalf_power)
y = statistics.mean(saruman_power)

print("The average spell power of Gandalf is:", x)
print("The average spell power of Saruman is:", y, "\n")


# Standard deviation of the spell power of Gandalf and Saruman.-----------------------------------------------------------------

import statistics

x = statistics.stdev(gandalf_power)
y = statistics.stdev(saruman_power)

print("The standard desviation of the spell power of Gandalf is:", x)
print("The standard desviation of the spell power of Saruman is:", y)

```
